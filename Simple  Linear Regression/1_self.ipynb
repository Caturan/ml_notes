## Median
* The median can be thought of as the middle values in the data set.
* To calculate the median, first need to sort the data. Then we finf the middle value of the sorted data. This represents the median. 


## Should we use the mean or the meadian to decribe the central tendency of the data? 
* In most cases we use the mean when the data show a symmetric distribution. 
* If we cut the data into two equal parts, the distribution looks about the same in the two regions. That is the symmetric distribution. 
* If we cut there is no equal parts that is the not symmetric. In this case we would say that distribution is skewed. 
* When the data points show a symmetric distribution, the mean and median will be very close. 
* However, for skewed distributions, the mean will be highly influenced by extreme values. 
* In contrast, the median is less sensitive to such extreme values because it is based on the middle value. 
* For skewed distributions, the median is a better measure of the central tendency because it shows the center of the majority of the data points.

# Spread of data 
* The spread of the data tells us how close or how far away the data points are from each other. 
* One measure of the spread of the data is the so-called range. The range of a data set is defined as the difference between the largest ad smalles value.

## Interquartile range 
* The interquartile range (IQR) is another measure of the spread of the data. The IQR is the difference between the 3rd quartile and the 1st quartile.
* Quartiles are values that divide ordered numbers into quarters.
* The interquartile range is defined as the range for the middle 50% of the ordered data.
A box plot, or box and whisker plot, is a graphical representation of the median, the quartiles and the range. 



## Standard deviation 

* Standard deviation is: 2 ; Mean is: 4 ; The standard deviation is usually presented in association with the mean like this. 4+-2 
* The standard deviation is usually represented graphically in bar charts.

* Where the height of the bar corresponds to the mean, and the so-called error bar corresponds to one standard deviation.

## Variance 

* The sample variance is related to the standard deviation.
* The variance is denoted as the square of s, simply because it is the square of the standard deviation. 
* Therefore, if we know the standard deviation, we can square it to get the variance. 
* The variance is mainly used in theoretical statistics.

## Which measure should we select to decribe the spread of the data? 
* Generally, use the standard deviation in association with the mean and the interquartile range with the median. 
* That one standard deviation away from the mean covers about %68 of data, given that the data has a normal distribution, and that the interval between quartile one and quartile three covers the middle %50 of the data points. 

### Why do we divide by n-1 and not n? 
* We divide by n minus one is because the data points are almost always closer to the sample mean compared to the population mean.
