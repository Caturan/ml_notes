# Transformers in Machine Learning

- Transformers is a neural network architecture used for performing machine learning tasks particularly in natural language processing (NLP) and computer vision. 

## Need For Transformers Model in Machine Learning

- Transformer architecture uses an attention mechanism to process an entire sentence at once instad of reading words one by one. 
- Older models work step by step. 
- Traditional models like RNNs suffor from the vanishing, gradient problem which leads to long-term memory loss. 
- RNNs process text sequentially meaning they analyze words one at a time. 

- While adding more memory cells in LSTMs (Long Short-Term Memory networks) helped adress the vanishing gradient issue they still process words one by one. 
- This sequential processing means LSTMs can't analyze an entire sentence at one. 

- Traditional models struggle with this context dependence, whereas Transformer model through its self-attention mechanism processes the entire sentence in parallel addressing these issues and making it significantly more effective at understanding context. 

## Core Concepts of Transformers

1. Self Attention Mechanism: 
    - The self attention mechanism allows transformers to determine which words in a sentence are most relevant to each other. 
    - This is done using a scaled dot-product attention approach. 
    - Each word in a sequence is mapped to three vectors: 
        - Query (Q)
        - Key (K)
        - Value (V)

2. Multi-Head Attention 
    - Instead of one attention mechanism, transformers use multiple attention heads running in parallel. 
    - Each head captures different relationships or patterns in the data, enriching the model's understanding. 

3. Positional Encoding 
    - Unlike RNNs, transformers lack and inherent understanding of word order since they process data in parallel. 
    - To solve this problem Positional Encoding are added to token embeddings providing information about the position of each token within a sequence. 

4. Position-wide Feed-Forward Networks 
    - The Feed-Forward Networks consist of two linear transformations with a ReLU activation. 

5. Embeddings 
    - Transformers cannot work with raw words as they need numbers. So, each input token is converted into a vector, called an embedding. 
    - Both encoder input tokens and decoder input tokens are converted into embeddings.
    - These embeddings are trainable, meaning the model learns the best numeric representation for each token.
    - The same weight matrix is shared for Encoder embeddings, Decoder embeddings and the final linear layer before softmax
    - The embeddings are scaled by model to keep values stable before adding positional encoding.
    - Embeddings turn words into meaningful numeric vectors that the transformer can process. 

6. Encoder-Decoder Architecture 
    - The encoder-decoder structure is key to transformer models. 
    - The encoder processes the input sequence into a vector, while the decoder converts this vector back into a sequence. 
    - Each encoder and decoder layer includes self-attention and feed-forward layers. 

7. Softmax Layer for Output Prediction 
    - After the decoder processes the sequence, it must predict the next token. 
        - The decoder output is passed through a linear layer. 
        - Then the softmax function converts these scores into probabilities. 
        - The token with the highest probality becomes the predicted next word.
