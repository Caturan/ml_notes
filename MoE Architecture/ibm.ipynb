# What is the MoE 

Mixture of experts (MoE) is a machine learning approach that divides an artificial intelligence (AI) model into separate sub-networks (or "experts"), each specializing in a subset of the input data, to jointly perform a task. 

- Mixture of Experts architectures enable large-scale models, even those comprising many billions of parameters, to greatly reduce computation costs during pre-training and achieve faster performance during inference time. 

## Mixtures of experts in deep learning 

- Modern deep learning models are built from artificial neural networks. Each neuran has an activation function: a mathematical operation performed on data received from the previous layer, whose output informs the input fed to the following layer. 

- A deep learning model "learns" by adjusting these parameters, using optimization algorithm like gradient descent, in a way that increases the accuracy of its predictions. 

- While a larger numbers of parameters increases the model's capacity -its ability to absorb information and patterns therein -it also increases the computational resources needed to train and operate the model. 

- Unkike conventional dense models, mixture of expers uses condiditional computation to enforce sparsity: rather than using the entire network for every input, MoE models learns a computationaly cheap mapping function that determines which portions of the network -in other words, which experts -are most effective to process a given input, like an individual token uses to represent a word or fragment in NLP tasks. 

- This allows the capacity of the model to be increased without a corresponding increase in the computational burden required to train and run it. 

## How do mixture of expers models work? 

- MoE models process data by designating a number of "experts", each its own sub-network within a larger neural network, and training a gating network (or router) to activate only the specific expert(s) best suited to a given input. 

- On an architectural level, this is achieved by replacing traditional, dense feed-forward network (FFN) layers with sparse MoE layers (or blocks). 

- For example, in Mistral’s Mixtral 8x7B language model, each layer is composed of 8 feedforward blocks—that is, experts—each of which has 7 billion parameters. For every token, at each layer, a router network selects two of those eight experts to process the data.  It then combines the outputs of those two experts and passes the result to the following layer. The specific experts selected by the router at a given layer may be different experts from those selected at the previous or next layer.
