#### Hill Descending
- Hill Descending is an optimization technique where we iteratively move toward values of a function, aiming to minimize it (oppesite to Hill Climbing, which maximizes the function). This approach is used in machine learning, particularly in optimization tasks. 
- Algorithm: 
    1. Initialize: Start at a random point x0 in the function's domain. 
    2. Calculate Gradient: Compute the gradient deltaf(x) of the function f(x) at the current position. 
    3. Step in Opposite Direction: Move in the opposite direction of the gradient by a small step size. 
    4. Iterate: Repeat until the stopping criterion is met.

#### True Function for Generalization 
- In machine learning, the true function represents the actual underlying relationship between input features and the target variable. 
- We aim to minimize the difference between our model's predictions and this true function to ensure that the model performs well on new data, thus achieving good generalization.
