## Define the following terms and provide an example for each: 
- Mean: The mean is the average of a set of numbers, calculated by dividing the sum of all values by the number of values. 
- Median: Middle value in a dataset when the values are sorted. 
- Mode: Value that appears most frequently in a dataset. 
- Variation: Measures the spread of a set of values by calculating the average of the squared differences from the mean. 
- Standard Deviation: Square root of the variance. 

## Explain the importance of each measure in the context of a dataset.
- Mean: Provides a central value for the dataset, allowing for quick insight into the overall tendency of the data. It can be influenced by outliers. 
- Median: Robust to outliers and gives a better indication of the central tendency when the data is skewed. 
- Mode: Useful in categorical data analysis. 
- Variance: Quantifies the degree of spread in the data, helping the assess risk and variability. A high variance indicates that data points are widely dispersed from the mean. 
- Standard Deviation: Standard deviation provides a measure of spread in the same unit as the data, making it more interpretable than variance.

### Discuss how you would evaluate the performance of a multiple regression model. 
1. R-squared (R^2): 
    * Measures the proportion of variance in the dependent variable explained by the independent variables. 
    * An R^2 value close to 1 indicates a good fit, while a values close to 0 suggests a poor fit. 
2. Adjusted R-Squared:
    * Adjust the R^2 value based on the number of predictors in the model. It penalizes adding unnecessary variables, providing a more accurate measuere of model performance. 
3. Mean Absolute Error (MAE): 
    * The average absolute difference between predicted and observed values. Lower MAE indicates beter predictive accuracy. 
4. Mean Squared Error (MSE): 
    * The average of the squared differences between predicted and actual values. 
5. Cross-Validation: 
    * Use techniques such as k-fold cross-validation to evaluate how well the model generalized to unseen data.

## Brute Force
* The brute force method involves trying many combinations of w0 and w1 to find the values that minimize the cost function. This is not the most efficient method, but it can be useful for understanding how a model's parameters affect the cost.


## Gradient Descent Method 
* Gradient descent is an iterative optimization algorithm that adjusts w0 and w1 in small steps to minimize the cost function. Unlike brute force, gradient descent can converge to an optimal solution much faster, especially for large datasets.
